Carregando o conjunto de dados Íris, criando um DataFrame com os dados, adicionando a coluna das espécies e exibindo um gráfico de dispersão das características 'sepal length (cm)' e 'sepal width (cm)', colorindo os pontos por espécie.

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_data['species'] = iris.target_names[iris.target]

sns.set_style('whitegrid')
facet = sns.FacetGrid(iris_data, hue='species', height=5) # Use 'species' for consistency
facet.map(plt.scatter, 'sepal length (cm)', 'sepal width (cm)').add_legend() # Use actual column names

plt.show()
```

criando e exibindo um gráfico de caixas que permite visualizar a distribuição do comprimento das pétalas para cada espécie de Íris no DataFrame.

```python
iris_data = pd.DataFrame({
    'Especies': ['setosa', 'versicolor', 'virginica', 'setosa', 'setosa',
                'versicolor', 'virginica', 'setosa', 'versicolor', 'virginica'],
    'ComprimentoPetalasCm': [1.4, 4.7, 5.6, 1.5, 1.3, 4.5, 5.8, 1.4, 4.2, 5.7]
})

sns.set_style('whitegrid')
sns.boxplot(x='Especies', y='ComprimentoPetalasCm', data=iris_data, palette="Set2")
plt.grid(True)
plt.show()
```

Criando uma matriz de gráficos de dispersão coloridos de acordo com a espécie da Íris e exibindo o gráfico. O resultado final é uma visualização que permite entender melhor as relações entre as diferentes características medidas nas três espécies de Íris.


```python
iris_data = sns.load_dataset('iris')

sns.set_style('whitegrid')
sns.pairplot(iris_data, hue='species', height=1.5)
plt.show()
```

criando um gráfico de dispersão 3D onde cada ponto representa uma amostra de flor, colorida de acordo com sua espécie. O gráfico mostra a relação entre o comprimento da sépala, a largura da sépala e o comprimento das pétalas das três espécies de Íris.

```python
iris_data = sns.load_dataset('iris')

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

species_colors = {'setosa': 'r', 'versicolor': 'g', 'virginica': 'b'}
for species, color in species_colors.items():
    species_data = iris_data[iris_data['species'] == species]
    ax.scatter(species_data['sepal_length'], species_data['sepal_width'], species_data['petal_length'], c=color, label=species)

ax.set_xlabel('Comprimento Sépala')
ax.set_ylabel('Largura da Sépala')
ax.set_zlabel('Comprimento das Pétalas')
ax.set_title('Relação entre Comprimento Sépala, Largura da Sépala e Comprimento das Pétalas')
ax.legend()

plt.show()
```

criando um gráfico de pizza que mostra a proporção de cada espécie de flor Íris presente no conjunto de dados.

```python
species = iris_data['species'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(species, labels=species.index, autopct='%1.3f%%', explode=[0, 0, 0])
plt.title('Proporção de Espécies no Conjunto de Dados Iris')
plt.show()
```

criando um gráfico de violino para cada uma das variáveis contínuas ('sepal_length', 'sepal_width', 'petal_length', 'petal_width') em relação à variável dependente ('species') do conjunto de dados Iris. Isso permite visualizar a distribuição das variáveis contínuas para cada espécie de flor Iris.

```python

continuous_var = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
dependent_var = ['species']
df = sns.load_dataset('iris')

for i in continuous_var:
    plt.figure(figsize=(10,5))
    sns.violinplot(x=dependent_var[0], y=i, data=df, palette='Set2')
    plt.ylabel(i)
    plt.xlabel(dependent_var[0])
    plt.title(dependent_var[0]+' vs '+i)
    plt.show()
```

Simplificação dos gráficos acima, o objetivo deste gráfico é mostrar a distribuição das medidas de comprimento e largura das sépalas e pétalas para as diferentes espécies de Íris, permitindo visualizar como essas características variam entre as espécies

```python
df = sns.load_dataset('iris')

continuous_vars = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(12, 6))
sns.violinplot(data=df[continuous_vars], palette='Set2')
plt.title('Distribuição das Variáveis Contínuas para as Espécies de Íris')
plt.xlabel('Variáveis Contínuas')
plt.ylabel('Valores')
plt.show()
```
